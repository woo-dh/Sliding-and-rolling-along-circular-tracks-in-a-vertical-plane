```python
# experiment_a_two_graphs.py

import math
import os
import pandas as pd
import matplotlib.pyplot as plt

g = 9.81
R = 0.0475
mu = 0.33

dt = 0.0005
total_time = 1.2
steps = int(total_time / dt)

vertical_drop_height = 0.27
vertical_friction_factor = 0.05

section = "vertical fall"
v = 0.0
vertical_y = vertical_drop_height
theta = -math.pi / 2

time_values = []
section_values = []
height_values = []
v_values = []
potential_energy_values = []
kinetic_energy_values = []
total_energy_values = []

for step in range(steps + 1):
    t = step * dt

    if section == "vertical fall":
        height = vertical_y + R
        potential_energy_per_mass = g * height
        kinetic_energy_per_mass = 0.5 * v ** 2
        total_energy_per_mass = potential_energy_per_mass + kinetic_energy_per_mass

        time_values.append(t)
        section_values.append(section)
        height_values.append(height)
        v_values.append(v)
        potential_energy_values.append(potential_energy_per_mass)
        kinetic_energy_values.append(kinetic_energy_per_mass)
        total_energy_values.append(total_energy_per_mass)

        acceleration = g * (1 - vertical_friction_factor)
        v = v + acceleration * dt
        vertical_y = vertical_y - v * dt

        if vertical_y <= 0:
            section = "curved section"
            vertical_y = 0.0
            theta = -math.pi / 2

    elif section == "curved section":
        height = R * (1 - math.cos(theta))
        normal_per_mass = g * math.cos(theta) + (v ** 2 / R)
        normal_per_mass = max(0.0, normal_per_mass)

        potential_energy_per_mass = g * height
        kinetic_energy_per_mass = 0.5 * v ** 2
        total_energy_per_mass = potential_energy_per_mass + kinetic_energy_per_mass

        time_values.append(t)
        section_values.append(section)
        height_values.append(height)
        v_values.append(v)
        potential_energy_values.append(potential_energy_per_mass)
        kinetic_energy_values.append(kinetic_energy_per_mass)
        total_energy_values.append(total_energy_per_mass)

        if abs(v) < 0.001 and theta > 0:
            break

        if v > 0:
            friction_direction = -1
        elif v < 0:
            friction_direction = 1
        else:
            friction_direction = 0

        gravity_acceleration = -g * math.sin(theta)
        friction_acceleration = friction_direction * mu * normal_per_mass
        acceleration = gravity_acceleration + friction_acceleration

        v = v + acceleration * dt
        theta = theta + (v / R) * dt

        if theta >= math.pi / 2 and v > 0:
            theta = math.pi / 2
            v = 0.0
            break

df = pd.DataFrame({
    "time": time_values,
    "section": section_values,
    "height": height_values,
    "speed": v_values,
    "potential_energy_per_mass": potential_energy_values,
    "kinetic_energy_per_mass": kinetic_energy_values,
    "total_energy_per_mass": total_energy_values
})

os.makedirs("data", exist_ok=True)
os.makedirs("figures", exist_ok=True)

df.to_csv("data/experiment_a_two_graphs_result.csv", index=False)

vertical_df = df[df["section"] == "vertical fall"]
curved_df = df[df["section"] == "curved section"]

curved_start_time = curved_df["time"].iloc[0]

plt.figure(figsize=(10, 5))
plt.plot(
    vertical_df["time"],
    vertical_df["speed"],
    linestyle="--",
    linewidth=1.5,
    alpha=0.5,
    label="Vertical fall section"
)
plt.plot(
    curved_df["time"],
    curved_df["speed"],
    linewidth=3.0,
    label="Curved section"
)
plt.axvline(
    x=curved_start_time,
    linestyle=":",
    linewidth=2,
    label="Curved section begins"
)
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.title("Experiment A: Speed Change of Metal Nut")
plt.legend()
plt.grid(True)
plt.savefig("figures/experiment_a_speed_graph.png", dpi=300, bbox_inches="tight")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(
    vertical_df["time"],
    vertical_df["potential_energy_per_mass"],
    linestyle="--",
    linewidth=1.2,
    alpha=0.5,
    label="Potential energy during vertical fall"
)
plt.plot(
    vertical_df["time"],
    vertical_df["kinetic_energy_per_mass"],
    linestyle="--",
    linewidth=1.2,
    alpha=0.5,
    label="Kinetic energy during vertical fall"
)
plt.plot(
    vertical_df["time"],
    vertical_df["total_energy_per_mass"],
    linestyle="--",
    linewidth=1.2,
    alpha=0.5,
    label="Total energy during vertical fall"
)
plt.plot(
    curved_df["time"],
    curved_df["potential_energy_per_mass"],
    linewidth=2.5,
    label="Potential energy in curved section"
)
plt.plot(
    curved_df["time"],
    curved_df["kinetic_energy_per_mass"],
    linewidth=2.5,
    label="Kinetic energy in curved section"
)
plt.plot(
    curved_df["time"],
    curved_df["total_energy_per_mass"],
    linewidth=3.0,
    label="Total energy in curved section"
)
plt.axvline(
    x=curved_start_time,
    linestyle=":",
    linewidth=2,
    label="Curved section begins"
)
plt.xlabel("Time (s)")
plt.ylabel("Energy per mass (J/kg)")
plt.title("Experiment A: Mechanical Energy Change of Metal Nut")
plt.legend()
plt.grid(True)
plt.savefig("figures/experiment_a_energy_graph.png", dpi=300, bbox_inches="tight")
plt.show()

print("CSV saved to data/experiment_a_two_graphs_result.csv")
print("Speed graph saved to figures/experiment_a_speed_graph.png")
print("Energy graph saved to figures/experiment_a_energy_graph.png")
```
