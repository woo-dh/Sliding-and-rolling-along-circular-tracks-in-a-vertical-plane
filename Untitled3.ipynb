```python
import numpy as np
import matplotlib.pyplot as plt

class TrackSimulation:
    def __init__(self, R=0.0475, mu=0.33, g=9.81):
        self.R = R      # Track radius (meters)
        self.mu = mu    # Coefficient of friction
        self.g = g      # Gravitational acceleration

    def run_sliding_experiment(self, v0=2.3, t0=0.04):
        """
        [Case 1: Sliding - Experiment 1]
        Simulation of a metal nut sliding along a wire track.
        Analyzes velocity drop due to curvature-induced friction.
        """
        dt = 0.001
        t, v, theta = t0, v0, 0.0
        results = {'t': [], 'v': [], 'energy': []}

        while v > 0 and t < 0.25:
            # Normal force N = mg*cos(theta) + m*v^2/R
            N_over_m = self.g * np.cos(theta) + (v**2 / self.R)
            # Acceleration dv/dt = -g*sin(theta) - mu*(N/m)
            dvdt = -self.g * np.sin(theta) - self.mu * N_over_m

            v += dvdt * dt
            theta += (v * dt) / self.R
            t += dt

            results['t'].append(t)
            results['v'].append(v)
            # Specific energy (Energy/mass)
            results['energy'].append(0.5 * v**2 + self.g * self.R * np.sin(theta))

        return results

    def run_rolling_experiment(self, beta=0.4, h_start=0.5):
        """
        [Case 2: Rolling Motion]
        Analyzes objects with rotational inertia (beta), such as a billiard ball.
        """
        # Theoretical velocity: v = sqrt(2gh / (1 + beta))
        heights = np.linspace(h_start, 0, 100)
        velocities = np.sqrt((2 * self.g * (h_start - heights)) / (1 + beta))
        return heights, velocities

    def analyze_loop_the_loop(self, beta_list=[0, 0.4]):
        """
        [Case 3: Loop-the-loop Condition]
        Calculates the minimum required height for different objects.
        """
        print("--- Loop-the-Loop Analysis ---")
        for beta in beta_list:
            # Theoretical minimum height: h = R * (2.5 + 0.5 * beta)
            h_min = self.R * (2.5 + 0.5 * beta)
            label = "Sliding (Nut)" if beta == 0 else "Rolling (Ball)"
            print(f"{label}: Minimum required height h = {h_min:.4f} m")

# --- Execution and Visualization ---
sim = TrackSimulation()

# 1. Simulate Sliding (Replicating Fig. 8a)
res_sliding = sim.run_sliding_experiment()

# 2. Compare Rolling Models (Sliding beta=0 vs. Billiard beta=0.4)
h_rolling, v_sliding = sim.run_rolling_experiment(beta=0)
h_rolling, v_billiard = sim.run_rolling_experiment(beta=0.4)

# 3. Plotting Results
plt.figure(figsize=(12, 5))

# Plot 1: Sliding Experiment (Time vs. Velocity)
plt.subplot(1, 2, 1)
plt.plot(res_sliding['t'], res_sliding['v'], 'r-', label='Nut Sliding (Exp 1)')
plt.title('Experiment 1: Nut Velocity Analysis')
plt.xlabel('Time (s)'); plt.ylabel('Velocity (m/s)')
plt.grid(True); plt.legend()

# Plot 2: Rolling Model Comparison (Height vs. Velocity)
plt.subplot(1, 2, 2)
plt.plot(h_rolling, v_sliding, 'k--', label='Pure Sliding (beta=0)')
plt.plot(h_rolling, v_billiard, 'b-', label='Billiard Rolling (beta=0.4)')
plt.gca().invert_xaxis() # Invert x-axis to show decreasing height
plt.title('Rolling Model: Effect of Shape Factor (Beta)')
plt.xlabel('Current Height (m)'); plt.ylabel('Velocity (m/s)')
plt.grid(True); plt.legend()

plt.tight_layout()
plt.show()

# 3. Output Loop-the-loop Analysis
sim.analyze_loop_the_loop()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class PhysicsExperimentSim:
    """
    Simulation of sliding and rolling objects on circular tracks.
    Based on the paper: "Sliding and Rolling along Circular Tracks in a Vertical Plane" (Rod Cross, 2023).
    """
    def __init__(self):
        self.g = 9.81
        self.R = 0.0475  # Radius of the track used in Experiment 1 (47.5 mm)
        self.dt = 0.001  # Time step for numerical integration

    def experiment_1_sliding(self, mu=0.33):
        """
        [Experiment 1: Sliding Nut on Wire]
        Analysis of Fig. 8: Modeling friction loss due to curvature-induced normal force.
        Eq: dv/dt = -g*sin(theta) - mu * (g*cos(theta) + v^2/R)
        """
        # Starting conditions from Fig. 8a
        t, v, theta = 0.04, 2.3, 0.0
        results = {'t': [], 'v': []}

        while v > 0 and t < 0.25:
            # Normal force: N/m = g*cos(theta) + v^2/R
            N_over_m = self.g * np.cos(theta) + (v**2 / self.R)

            # Acceleration with friction dissipation
            dvdt = -self.g * np.sin(theta) - mu * N_over_m

            # Update state using Euler method
            v += dvdt * self.dt
            theta += (v * self.dt) / self.R
            t += self.dt

            results['t'].append(t)
            results['v'].append(v)

        return results

    def experiment_2_rolling(self, beta=0.4):
        """
        [Experiment 2: Rolling Billiard Ball]
        Simulation of a rolling object considering rotational inertia (beta).
        Eq: v = sqrt( (2 * g * h) / (1 + beta) )
        """
        h_start = 0.165  # Example drop height for the billiard ball experiment
        heights = np.linspace(h_start, 0, 100)

        # Velocity calculation based on energy conservation including rotation
        v_rolling = np.sqrt((2 * self.g * (h_start - heights)) / (1 + beta))

        return heights, v_rolling

    def simulation_loop_the_loop(self):
        """
        [Scenario 3: Loop-the-Loop Simulation]
        Comparative analysis of minimum release height (h_min) for different objects.
        Eq: h_min = R * (2.5 + 0.5 * beta)
        """
        objects = {
            'Sliding Nut (beta=0)': 0,
            'Billiard Ball (beta=0.4)': 0.4,
            'Hollow Sphere (beta=0.67)': 0.67
        }
        loop_results = {}
        for name, b in objects.items():
            h_min = self.R * (2.5 + 0.5 * b)
            loop_results[name] = h_min
        return loop_results

# --- Execution and Visualization ---
exp = PhysicsExperimentSim()

plt.figure(figsize=(15, 5))

# Plot 1: Sliding Nut Analysis (Experiment 1)
res1 = exp.experiment_1_sliding()
plt.subplot(1, 3, 1)
plt.plot(res1['t'], res1['v'], 'r-', linewidth=2, label='Sliding Nut (Exp 1)')
plt.title('Exp 1: Velocity vs Time (Sliding)')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.grid(True, alpha=0.3)
plt.legend()

# Plot 2: Rolling Ball Analysis (Experiment 2)
h, v_roll = exp.experiment_2_rolling()
plt.subplot(1, 3, 2)
plt.plot(h, v_roll, 'b-', linewidth=2, label='Billiard Ball (beta=0.4)')
plt.gca().invert_xaxis() # Show descent from left to right
plt.title('Exp 2: Velocity vs Height (Rolling)')
plt.xlabel('Height (m)')
plt.ylabel('Velocity (m/s)')
plt.grid(True, alpha=0.3)
plt.legend()

# Plot 3: Loop-the-Loop Comparison
loop_data = exp.simulation_loop_the_loop()
plt.subplot(1, 3, 3)
names = list(loop_data.keys())
heights = list(loop_data.values())
bars = plt.bar(names, heights, color=['orange', 'green', 'purple'])
plt.axhline(y=exp.R*2, color='black', linestyle='--', label='Loop Top (2R)')
plt.title('Minimum Height for Loop Completion')
plt.ylabel('Height (m)')
plt.xticks(rotation=15)
plt.legend()

plt.tight_layout()
plt.show()

# Print Numerical Results
print("--- Loop-the-Loop Threshold Analysis ---")
for obj, h_min in loop_data.items():
    print(f"{obj}: h_min = {h_min:.4f} m")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Case 1: Sliding Nut on Wire (Exp 1) ---
g = 9.81
R_track = 0.0475
mu_sliding = 0.33
dt = 0.001

v = 2.3  # Initial velocity at curve entry (t=0.04s)
theta = 0.0
results_sliding = []

# Time loop for the curved section
for i in range(210):  # Roughly until the nut stops (approx 0.19s - 0.20s)
    t = 0.04 + (i * dt)

    # Physics: Normal force includes gravity and centripetal components
    normal_force_per_m = g * np.cos(theta) + (v**2 / R_track)

    # Acceleration: Gravity opposes motion as it goes up, friction always opposes
    acceleration = -g * np.sin(theta) - mu_sliding * normal_force_per_m

    v = max(0.0, v + acceleration * dt)
    theta += (v * dt) / R_track

    results_sliding.append({
        "time": t,
        "velocity": v,
        "theta_deg": np.degrees(theta)
    })
    if v <= 0: break

df_sliding = pd.DataFrame(results_sliding)
```

```python
# --- Case 2: Rolling Billiard Ball (Exp 2) ---
r_ball = 0.0254
r_eff = 0.0230  # Effective rolling radius on track
beta = 0.4      # Shape factor for solid sphere
v_roll = 0.0
omega_roll = 0.0
results_rolling = []

for i in range(1600):
    t_roll = i * dt

    # Section 1: 60-degree Incline Descent (Sliding to Rolling transition)
    if t_roll < 0.26:
        acc = 6.61  # Measured acceleration from paper
        v_roll += acc * dt
        # Torque is limited; omega increases slower than pure rolling
        omega_roll += (0.35 * acc / r_eff) * dt

    # Section 2: First Curved Section (Friction increases rotation)
    elif t_roll < 0.35:
        v_roll *= 0.992  # Friction loss
        omega_roll += 45.0 * dt  # Rapid rotation increase due to track friction

    # Section 3: Up 30-degree Incline (Approaching pure rolling)
    else:
        acc = -g * np.sin(np.radians(30)) * 0.55
        v_roll = max(0.0, v_roll + acc * dt)
        omega_roll = v_roll / r_eff  # Becomes pure rolling

    results_rolling.append({
        "time": t_roll,
        "v": v_roll,
        "omega": omega_roll,
        "v_omega_ratio": v_roll / omega_roll if omega_roll > 0 else 0
    })

df_rolling = pd.DataFrame(results_rolling)
```

```python
# --- Case 3: Loop-the-Loop Thresholds ---
objects = {
    "Sliding Nut": 0.0,
    "Billiard Ball": 0.4,
    "Hollow Sphere": 0.67
}

loop_data = []
for name, b in objects.items():
    # Theoretical h_min = R * (2.5 + 0.5 * beta)
    h_min = R_track * (2.5 + 0.5 * b)
    loop_data.append({"Object": name, "Min_Height": h_min})

df_loop = pd.DataFrame(loop_data)
```
